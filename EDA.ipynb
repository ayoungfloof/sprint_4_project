```python
import pandas as pd
import plotly.express as px
from ipywidgets import widgets

# Load the dataset
df = pd.read_csv('vehicles_us.csv')

# Create the 'manufacturer' column by extracting the first word from the 'model' column
df['manufacturer'] = df['model'].str.split().str[0]

# Check for NaN values and remove them in the 'days_listed' column
df = df.dropna(subset=['days_listed'])

# Optionally, you can limit the number of rows to test with a smaller dataset
df_sample = df.head(10000)  # Change this as needed for testing smaller portions of the data

# Create dropdowns for selecting up to three manufacturers
manufacturers = df['manufacturer'].unique()
dropdown_1 = widgets.Dropdown(options=manufacturers, description="Manufacturer 1")
dropdown_2 = widgets.Dropdown(options=manufacturers, description="Manufacturer 2")
dropdown_3 = widgets.Dropdown(options=manufacturers, description="Manufacturer 3")

# Define a function to update the plot based on selected manufacturers
def update_plot(manufacturer_1, manufacturer_2=None, manufacturer_3=None):
    # Filter data based on selected manufacturers
    selected_manufacturers = [manufacturer_1]
    if manufacturer_2: selected_manufacturers.append(manufacturer_2)
    if manufacturer_3: selected_manufacturers.append(manufacturer_3)
    
    # For testing, use the df_sample (smaller portion of the data) to speed up the process
    filtered_df = df_sample[df_sample['manufacturer'].isin(selected_manufacturers)]
    
    # Create the histogram plot
    fig = px.histogram(
        filtered_df,
        x='days_listed',  # Display distribution of days listed
        color='manufacturer',  # Color by manufacturer
        title=f'Distribution of Days Listed by Selected Manufacturers',
        labels={'days_listed': 'Days Listed', 'count': 'Number of Vehicles'},  # Update y-axis label
        barmode='overlay',  # Overlay the histograms for comparison
        nbins=20  # Set a smaller number of bins for faster performance
    )
    
    # Update the layout to add horizontal grid lines and customize the y-axis label
    fig.update_layout(
        yaxis=dict(title='Number of Vehicles', showgrid=True, gridcolor='lightgray', gridwidth=1),  # Updated y-axis label
        xaxis=dict(title='Days Listed', showgrid=False)  # Optional: remove vertical grid lines
    )
    
    # Show the plot
    fig.show()

# Connect the dropdown selections with the update_plot function
widgets.interactive(update_plot, manufacturer_1=dropdown_1, manufacturer_2=dropdown_2, manufacturer_3=dropdown_3)
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import widgets, interactive_output
from IPython.display import display

# Bin the odometer readings into ranges
bins = [0, 25000, 50000, 75000, 100000, 150000, 200000, 250000, 300000]
labels = ['0-25K', '25K-50K', '50K-75K', '75K-100K', '100K-150K', '150K-200K', '200K-250K', '250K+']
df['odometer_range'] = pd.cut(df['odometer'], bins=bins, labels=labels)

# Create dropdowns for selecting three manufacturers
manufacturers = df['manufacturer'].unique()
dropdown_1 = widgets.Dropdown(options=manufacturers, description="Manufacturer 1")
dropdown_2 = widgets.Dropdown(options=manufacturers, description="Manufacturer 2")
dropdown_3 = widgets.Dropdown(options=manufacturers, description="Manufacturer 3")

# Define a function to update the histogram plot based on selected manufacturers
def update_plot(manufacturer_1, manufacturer_2=None, manufacturer_3=None):
    selected_manufacturers = [manufacturer_1]
    if manufacturer_2: selected_manufacturers.append(manufacturer_2)
    if manufacturer_3: selected_manufacturers.append(manufacturer_3)
    
    # Efficiently filter the data using .loc and only keep the necessary columns
    filtered_df = df.loc[df['manufacturer'].isin(selected_manufacturers), ['manufacturer', 'odometer_range', 'price']]
    
    # Group the filtered data by 'manufacturer' and 'odometer_range' and calculate the average price
    avg_price_by_odometer_range = filtered_df.groupby(['manufacturer', 'odometer_range'])['price'].mean().reset_index()

    # Create the histogram to compare average price by odometer range
    fig = px.histogram(
        avg_price_by_odometer_range,
        x='odometer_range',
        y='price',  # Compare by average price
        color='manufacturer',  # Color by manufacturer
        barmode='overlay',  # Overlay bars
        title=f'Average Price by Odometer Range for Selected Manufacturers',
        labels={'price': 'Average Price (USD)', 'odometer_range': 'Odometer Range (Miles)'}
    )
    
    # Update layout to add grid lines and improve readability
    fig.update_layout(
        yaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1),  # Customize grid lines
        xaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1)  # Add grid lines for x-axis
    )
    
    # Show the histogram plot
    fig.show()

# Connect the dropdown selections with the update_plot function
out = interactive_output(update_plot, {'manufacturer_1': dropdown_1, 'manufacturer_2': dropdown_2, 'manufacturer_3': dropdown_3})

# Display dropdowns and the histogram plot
display(dropdown_1, dropdown_2, dropdown_3, out)
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import widgets, interactive_output
from IPython.display import display

# Create a scatter plot comparing price vs. odometer without summing the odometer
def update_scatter_plot(vehicle_condition=None):
    # Filter data by the selected vehicle condition if provided
    if vehicle_condition:
        filtered_df = df.loc[df['condition'] == vehicle_condition, ['odometer', 'price', 'manufacturer', 'model', 'condition']]
    else:
        filtered_df = df[['odometer', 'price', 'manufacturer', 'model', 'condition']]  # Select only necessary columns

    # Create the scatter plot
    fig = px.scatter(
        filtered_df,
        x='odometer',
        y='price',
        color='condition',  # Color by vehicle condition
        title=f'Price vs. Odometer Reading for {vehicle_condition or "All Conditions"}',
        labels={'price': 'Price (USD)', 'odometer': 'Odometer Reading (Thousands of Miles)'},  # Update label
        hover_data=['manufacturer', 'model']  # Display manufacturer and model on hover
    )
    
    # Update layout for better readability and 'k' labels on x-axis
    fig.update_layout(
        xaxis=dict(
            title='Odometer (Thousands of Miles)',
            showgrid=True, 
            gridcolor='lightgray',
            tickvals=[0, 50000, 100000, 150000, 200000, 250000, 300000],  # Define tick values
            ticktext=['0k', '50k', '100k', '150k', '200k', '250k', '300k']  # Custom text with 'k'
        ),
        yaxis=dict(title='Price (USD)', showgrid=True, gridcolor='lightgray'),
        plot_bgcolor='white'
    )
    
    # Show the scatter plot
    fig.show()

# Create a dropdown to filter by vehicle condition
vehicle_conditions = df['condition'].unique()
dropdown_condition = widgets.Dropdown(options=vehicle_conditions, description="Condition")

# Display dropdown and scatter plot
out = interactive_output(update_scatter_plot, {'vehicle_condition': dropdown_condition})
display(dropdown_condition, out)
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import widgets, interactive_output
from IPython.display import display

# Bin the odometer readings into ranges for comparison
bins = [0, 25000, 50000, 75000, 100000, 150000, 200000, 250000, 300000]
labels = ['0-25K', '25K-50K', '50K-75K', '75K-100K', '100K-150K', '150K-200K', '200K-250K', '250K+']
df['odometer_range'] = pd.cut(df['odometer'], bins=bins, labels=labels)

# Get unique manufacturers and vehicle types for dropdown options
manufacturers = df['manufacturer'].unique()
vehicle_types = df['type'].unique()

# Create dropdowns for selecting manufacturer and type for three options
dropdown_manufacturer_1 = widgets.Dropdown(options=manufacturers, description="Manufacturer 1")
dropdown_type_1 = widgets.Dropdown(options=vehicle_types, description="Vehicle Type 1")

dropdown_manufacturer_2 = widgets.Dropdown(options=manufacturers, description="Manufacturer 2")
dropdown_type_2 = widgets.Dropdown(options=vehicle_types, description="Vehicle Type 2")

dropdown_manufacturer_3 = widgets.Dropdown(options=manufacturers, description="Manufacturer 3")
dropdown_type_3 = widgets.Dropdown(options=vehicle_types, description="Vehicle Type 3")

# Define a function to update the plot based on selected manufacturers and types
def update_plot(manufacturer_1, vehicle_type_1, manufacturer_2, vehicle_type_2, manufacturer_3, vehicle_type_3):
    # Filter data based on selected manufacturers and types
    selected_df = df[
        ((df['manufacturer'] == manufacturer_1) & (df['type'] == vehicle_type_1)) |
        ((df['manufacturer'] == manufacturer_2) & (df['type'] == vehicle_type_2)) |
        ((df['manufacturer'] == manufacturer_3) & (df['type'] == vehicle_type_3))
    ][['manufacturer', 'type', 'odometer_range', 'price']]

    # Group by 'manufacturer', 'type', and 'odometer_range' after filtering and calculate the average price
    avg_price_by_manufacturer_type_odometer = selected_df.groupby(['manufacturer', 'type', 'odometer_range'])['price'].mean().reset_index()

    # Create a new column to reflect both manufacturer and type in the legend
    avg_price_by_manufacturer_type_odometer['manufacturer_type'] = avg_price_by_manufacturer_type_odometer['manufacturer'] + " (" + avg_price_by_manufacturer_type_odometer['type'] + ")"

    # Create the bar plot color-coded by manufacturer_type (combining manufacturer and type in legend)
    fig = px.bar(
        avg_price_by_manufacturer_type_odometer,
        x='odometer_range',
        y='price',  # Compare by price
        color='manufacturer_type',  # Color by combined manufacturer and type
        barmode='group',  # Group bars side by side
        title=f'Average Price by Odometer Range for Selected Manufacturers and Vehicle Types',
        labels={'price': 'Average Price (USD)', 'odometer_range': 'Odometer Range (Miles)'}
    )
    
    # Update layout to add grid lines
    fig.update_layout(
        yaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1),  # Customize grid lines
        xaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1),  # Add grid lines for x-axis
        legend_title_text='Manufacturer and Vehicle Type'  # Customize legend title
    )
    
    # Show the plot
    fig.show()

# Connect the dropdown selections with the update_plot function
out = interactive_output(update_plot, {
    'manufacturer_1': dropdown_manufacturer_1, 'vehicle_type_1': dropdown_type_1,
    'manufacturer_2': dropdown_manufacturer_2, 'vehicle_type_2': dropdown_type_2,
    'manufacturer_3': dropdown_manufacturer_3, 'vehicle_type_3': dropdown_type_3
})

# Display dropdowns and graph
display(dropdown_manufacturer_1, dropdown_type_1,
        dropdown_manufacturer_2, dropdown_type_2,
        dropdown_manufacturer_3, dropdown_type_3, out)
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import widgets, interactive_output, Layout
from IPython.display import display

# Extract the first word from the 'model' column to create the 'manufacturer' column if it's not already created
if 'manufacturer' not in df.columns:
    df['manufacturer'] = df['model'].str.split().str[0]

# Create dropdowns for selecting vehicle type and manufacturer
manufacturers = df['manufacturer'].unique()
vehicle_types = df['type'].unique()

dropdown_manufacturer = widgets.Dropdown(options=manufacturers, description="Manufacturer")
dropdown_type = widgets.Dropdown(options=vehicle_types, description="Vehicle Type")

# Create range slider for selecting model year range (limited to 1929 onwards)
slider_model_year_range = widgets.IntRangeSlider(
    value=[2000, 2024],  # Default range
    min=1929, max=int(df['model_year'].max()),  # Limit to 1929 onwards
    step=1,
    description='Model Year Range',
    continuous_update=False,
    orientation='horizontal', 
    readout=True,
    readout_format='d',
    layout=Layout(width='80%')  # Make slider longer for easier selection
)

# Define a function to update the plot based on selected vehicle type, manufacturer, and model year range
def update_plot(vehicle_type, manufacturer, model_year_range):
    # Filter data based on selected vehicle type, manufacturer, and model year range
    filtered_df = df.loc[
        (df['type'] == vehicle_type) & 
        (df['manufacturer'] == manufacturer) & 
        (df['model_year'] >= model_year_range[0]) & 
        (df['model_year'] <= model_year_range[1]),
        ['model_year', 'price']  # Minimize columns to only what's needed
    ]
    
    # Group by 'model_year' and calculate the average price
    avg_price_by_year = filtered_df.groupby(['model_year'])['price'].mean().reset_index()

    # Create a bar plot to compare average price by model year for the selected vehicle type and manufacturer
    fig = px.bar(
        avg_price_by_year,
        x='model_year',
        y='price',
        title=f'Average Price vs. Model Year for {manufacturer} {vehicle_type}',
        labels={'price': 'Average Price (USD)', 'model_year': 'Model Year'}
    )
    
    # Customize layout
    fig.update_layout(
        yaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1),
        xaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1),
        showlegend=False  # Only one vehicle type is selected, so no need for a legend
    )
    
    # Show the plot
    fig.show()

# Display dropdowns, slider, and graph
out = interactive_output(update_plot, {'vehicle_type': dropdown_type, 'manufacturer': dropdown_manufacturer, 'model_year_range': slider_model_year_range})
display(dropdown_manufacturer, dropdown_type, slider_model_year_range, out)
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import widgets, interactive_output, Layout
from IPython.display import display

# Get unique vehicle types from the dataset
vehicle_types = df['type'].unique()

# Create dropdown for selecting multiple vehicle types
dropdown_type = widgets.SelectMultiple(
    options=vehicle_types,
    value=[vehicle_types[0]],  # Default to the first vehicle type
    description='Vehicle Types',
    layout=Layout(width='50%'),  # Make dropdown wider for easier selection
)

# Define a function to update the plot based on selected vehicle types
def update_plot(selected_vehicle_types):
    # Filter data based on selected vehicle types
    filtered_df = df.loc[df['type'].isin(selected_vehicle_types), ['type', 'transmission', 'price']]  # Minimize columns to needed ones
    
    # Group by 'type' and 'transmission', calculate average price
    avg_price_by_transmission = filtered_df.groupby(['type', 'transmission'])['price'].mean().reset_index()

    # Create a bar plot to compare price by transmission for selected vehicle types
    fig = px.bar(
        avg_price_by_transmission,
        x='type',
        y='price',  # Compare by average price
        color='transmission',  # Color by transmission type (automatic or manual)
        barmode='group',  # Group bars side by side
        title=f'Average Price by Vehicle Type and Transmission Type',
        labels={'price': 'Average Price (USD)', 'type': 'Vehicle Type', 'transmission': 'Transmission'},
        color_discrete_sequence=px.colors.qualitative.Vivid  # Custom color palette
    )
    
    # Customize layout with hover effects, title, font sizes, gridlines, and color palette
    fig.update_traces(
        hovertemplate='<b>%{x}</b><br>Transmission: %{marker.color}<br>Price: $%{y:,.0f}',  # Hover effect
    )
    fig.update_layout(
        title=dict(text='Average Price by Vehicle Type and Transmission Type', x=0.5, font=dict(size=20)),
        xaxis=dict(title='Vehicle Type', tickangle=-45, titlefont=dict(size=16), showgrid=False),
        yaxis=dict(title='Average Price (USD)', titlefont=dict(size=16), showgrid=True, gridcolor='lightgray', gridwidth=1),
        legend_title_text='Transmission Type',  # Title for the legend
        font=dict(size=12),  # General font size
        plot_bgcolor='white'  # Background color
    )
    
    # Show the plot
    fig.show()

# Connect dropdown and the plot function
out = interactive_output(update_plot, {'selected_vehicle_types': dropdown_type})

# Display dropdown and the graph
display(dropdown_type, out)
```
